```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


```

# Part 1

```python


text=("Peko Nichols Farrel Butler Billings Tilery Rochell Koonce Ferrell Crosby Bower")

# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=10).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

I decided to choose the Wordcloud, because most of the other visualizations were about numbers. And I wanted to try a new visualization that I am not very familiar with. In this case, the data I am plotting are the last names of the defensive line players of the Las Vegas Raiders football team.
