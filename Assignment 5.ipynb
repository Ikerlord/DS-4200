```python
import pandas as pd
import altair as alt
india_df = pd.read_csv("India_GDP_1960-2022.csv")
india_df
```

# India GDP and GDP Per Capita by Year Line Plot

```python
import altair as alt

chart56 = alt.Chart(india_df).mark_line(interpolate='step-after').encode(
    x=alt.X('Year', title='Year'),
    y=alt.Y('GDP in USD$', title='GDP in USD$'),
    color=alt.value("orange"),
    tooltip=['Year', alt.Tooltip('GDP in USD$', format='$,.2f')]
).properties(
    width=800,
    height=400,
    title={
        "text": "India GDP and GDP Per Capita by Year",
        "subtitle": "Yellow line = GDP, Red line = GDP Per Capita"
    }
)

chart2 = alt.Chart(india_df).mark_line(interpolate='step-after').encode(
    x=alt.X('Year', title='Year'),
    y=alt.Y('Per Capita in USD', title='Per Capita in USD'),
    color=alt.value("Red"),
    tooltip=['Year', alt.Tooltip('Per Capita in USD', format='$,.2f')]
).properties(
    width=800,
    height=400,
    title={
        "text": "India GDP by Year",
        "subtitle": "Yellow line = GDP, Red line = GDP Per Capita"
    }
)

chart56 + chart2


```

This graph is a line plot that shows India GDP and its GDP Per Capita through time. The yellow line represents the GDP, and the red one shows GDP per Capita. I decided to use colors in order to let the user distinguish between each line. The legend does not work, for that reason I decided to include the line indication in the title and in the description. On another note, the red line looks as it is straight, this is due to the fact that the GDP Per Capita is a very small number compared to the GDP. That is another reason why I decided to include the tooltip and it is because you can observe the number if you hover over the line. These tooltip solves the issue that GDP Per Capita looks irrelevant.

# Boxplot of India's GDP from 1960 to 2021

```python
brush = alt.selection_interval()

boxplot = alt.Chart(india_df).mark_boxplot(extent='min-max').encode(
    x=alt.X('Year', title='Year', scale=alt.Scale(domain=(1960, 2021))),
    y=alt.Y('GDP in USD$', title='GDP in USD$')
).properties(
    title='Boxplot of India\'s GDP from 1960 to 2021'
).add_selection(
    brush
)

scatter = alt.Chart(india_df).mark_circle().encode(
    x=alt.X('Year', title='Year', scale=alt.Scale(domain=(1960, 2021))),
    y=alt.Y('GDP in USD$', title='GDP in USD$'),
    color=alt.condition(brush, 'GDP in USD$', alt.value('lightgray')),
    tooltip=[alt.Tooltip('Year', title='Year'), alt.Tooltip('GDP in USD$', title='GDP in USD$')]
).properties(
    title='Scatter plot of India\'s GDP from 1960 to 2021',
    width=800,
    height=400
).add_selection(
    brush
)

boxplot = boxplot + alt.Chart(india_df).mark_rule().encode(
    x='Year',
    y='median(GDP in USD$)'
)

chart = (scatter | boxplot).resolve_scale(color='independent')

chart



```

This boxplot has a tooltip that is going to tell the information if you scroll through each line. I decided to use this dynamic tool because the user will be able to interact with the graph and learn from it. I also have a brush tool that allows you to look at specific data you want to look at.

# Boxplot of India's GDP from 1960 to 2021 With Slider

```python
year_slider = alt.binding_range(min=1960, max=2021, step=1, name='Year:')
year_select = alt.selection_single(name="year", fields=['Year'], bind=year_slider, init={'Year': 1960})

chart_3 = alt.Chart(india_df).mark_boxplot(extent='min-max').encode(
    x=alt.X('Year', title='Year', scale=alt.Scale(domain=(1960, 2025))),
    y=alt.Y('GDP in USD$', title='GDP in USD$')
).properties(
    title='Boxplot of India\'s GDP from 1960 to 2021'
).add_selection(
    year_select
).transform_filter(
    year_select
)

scatter = alt.Chart(india_df).mark_circle().encode(
    x=alt.X('Year', title='Year', scale=alt.Scale(domain=(1960, 2021))),
    y=alt.Y('GDP in USD$', title='GDP in USD$'),
    color=alt.condition(year_select, 'GDP in USD$', alt.value('lightgray')),
    tooltip=[alt.Tooltip('Year', title='Year'), alt.Tooltip('GDP in USD$', title='GDP in USD$')]
).properties(
    width=800,
    height=400,
    title=chart_3.title
).add_selection(
    year_select
).transform_filter(
    year_select
)

chart_3 = chart_3 + alt.Chart(india_df).mark_rule().encode(
    x='Year',
    y='median(GDP in USD$)'
).transform_filter(
    year_select
).properties(
    title=chart_3.title
)

chart_3 = (scatter | chart_3).resolve_scale(color='independent')

chart_3

```

The slider is important it allows the user to interact, select, and filter the viz based on given values. The users have a better understanding of trends and patterns in the viz The use of this slider will lead to a more engaging and informative user experience.
