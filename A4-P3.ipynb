```python
import pandas as pd
df_3 = pd.read_excel("table3 (2).xlsx")
df_3
```

# Table 3

```python
import altair as alt


alt.Chart(df_3).mark_bar().encode(
    x='Land cover types:O',
    y='Area',
    color='Land cover types:N',
    column='Location'
)
```

This graph is providing a clear understanfing of the Area within each location, and it is also breaking the area by differnt categories. In the Charles and Neponset we can note that most of the area is made up of forests. This charts are going to allow the user to understand how the area in a location is broken up.
